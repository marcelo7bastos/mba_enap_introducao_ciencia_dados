```python
project_id = 'starry-fiber-470814-p6' # Defina o id do seu projeto no bigquery!!!!!
## Defina o id do seu projeto no bigquery!!!!!

df = pd.io.gbq.read_gbq('''
SELECT * FROM
(SELECT pibtab.*, ideb.*
  FROM `enapdatasets.pibpercapita`  as pibtab
  INNER JOIN (

    WITH base AS (
      SELECT
        ano,
        idm.id_municipio,
        CONCAT(rede, '_', anos_escolares) AS rede_anos,
        nota_saeb_media_padronizada
      FROM `basedosdados.br_inep_ideb.municipio` idm
    )
    SELECT *
    FROM base
    PIVOT (
      AVG(nota_saeb_media_padronizada) FOR rede_anos IN (
        'municipal_finais (6-9)'    AS municipal_finais_6_9,
        'municipal_iniciais (1-5)'  AS municipal_iniciais_1_5,
        'municipal_todos (1-4)'     AS municipal_todos_1_4,
        'publica_finais (6-9)'      AS publica_finais_6_9,
        'publica_iniciais (1-5)'    AS publica_iniciais_1_5,
        'publica_todos (1-4)'       AS publica_todos_1_4,
        'estadual_finais (6-9)'     AS estadual_finais_6_9,
        'estadual_iniciais (1-5)'   AS estadual_iniciais_1_5,
        'estadual_todos (1-4)'      AS estadual_todos_1_4
      )
    )
    ORDER BY ano, id_municipio

) as ideb
  ON ideb.ano = pibtab.ano and ideb.id_municipio = pibtab.id_municipio
) as result
''', project_id=project_id)
```
